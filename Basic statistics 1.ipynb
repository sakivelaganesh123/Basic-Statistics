```python
# Step-1: Load the dataset into a data analysis tool or programming environment (e.g., Python with pandas library).
import pandas as pd
import statistics as st
```

```python
df = pd.read_csv("sales_data_with_discounts.csv")
df
```

```python
# Step-2: Identify numerical columns
num_df = df.select_dtypes(include=['number'])
print(num_df)
```

```python
num_df.columns
```

```python
# Step-3: Calculate the mean, median, mode, and standard deviation for these columns.
mean_values = num_df.mean()
print("Mean:\n", mean_values)

median_values = num_df.median()
print("\nMedian:\n", median_values)

mode_values = num_df.mode().iloc[0]
print("\nMode:\n", mode_values)

std_values = num_df.std()
print("\nStandard Deviation:\n", std_values)
```

```python
# Step-4: Provide a brief interpretation of these statistics.

# Mean Interpretation: A higher mean indicates a generally higher value across that column.

# Median Interpretation: A big gap between the mean and median may indicate outliers or skewed data.

# Mode Interpretation: A high-frequency mode may indicate a popular or standard value in that dataset.

# Standard Derivation Interpretation: If the standard deviation of salary is high, employee salaries vary a lot.
                                    # If it's low, salaries are fairly consistent.
```

```python
# Objective: To scale numerical variables for uniformity, improving the datasetâ€™s suitability for analytical models.
```

```python
# step-1: Explain the concept of standardization (z-score normalization).
```

```python
# Concept: Z-Score Normalization:
# Standardization (also called Z-score normalization) transforms your data using the following formula:
        # Standardization (z) = Original value(x) âˆ’ Mean of the feature(Î¼) / Standard deviation of the feature(Ïƒ)
# Where:
# x: Original value
# Î¼: Mean of the feature
# Ïƒ: Standard deviation of the feature
# z: Standardized value

# This ensures:
# The mean of the feature becomes 0
# The standard deviation becomes 1
```

```python
# step-2: Standardize the numerical columns using the formula: z=x-mu/sigma
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df.describe()
```

```python
scaler = StandardScaler()
scaled_arr = scaler.fit_transform(num_df)
```
